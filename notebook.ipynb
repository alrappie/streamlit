## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data Wrangling

### Gathering Data

```python
order_items_data = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\order_items_dataset.csv')
products_data = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\products_dataset.csv')
product_translation = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\product_category_name_translation.csv')
```

```python
order_items_data.head()
```

```python
products_data.head()
```

```python
product_translation.head()
```

```python
merged_data = pd.merge(order_items_data, products_data, how = 'inner',left_on= 'product_id',right_on= 'product_id')
merged_data.head()
```

```python
merged_data = pd.merge(merged_data, product_translation, how = 'inner',left_on= 'product_category_name',right_on= 'product_category_name')
merged_data.head()
```

**Insight:**
- Berdasarkan 2 pertanyaan diatas, dapat dijawab dengan menggunakan 3 table tersebut yaitu order_items_dataset untuk melihat produk yang dibeli, products_dataset untuk melihat nama produk, dan translation table untuk melihat produk dalam bahasa inggris. 
- Dikarenakan kedua product memiliki key yang sama yaitu product_id maka product_id dari masing-masing table akan dihubungkan kemudian akan dihubungkan berdasarkan category_name

### Cleaning Data

```python
data_cleaned = merged_data.dropna()
data_cleaned
```

**Insight:**
- dikarenakan hanya 4 data saja yang missing, maka dapat dilakukan dropna

## Exploratory Data Analysis (EDA)

### Explore ...

```python
data_cleaned.describe()
```

```python
most_selling_product = data_cleaned.groupby('product_category_name_english').agg({
    'product_id':'count',
    'price':'sum'
}).sort_values('product_id',ascending=False)
most_selling_product
```

```python
most_profitable_product = data_cleaned.groupby('product_category_name_english').agg({
    'product_id':'count',
    'price':'sum'
}).sort_values('price',ascending=False)
most_profitable_product
```

**Insight:**
- berdasarkan data tersebut category produk yang paling dibeli adalah bed_bath_table
- berdasarkan data tersebut category produk yang paling menguntungkan adalah health_beauty 

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import matplotlib.pyplot as plt

# Sort the products by sales in descending order
sorted_products = most_selling_product.sort_values(by='product_id', ascending=False).head(10)

# Create a horizontal bar plot
plt.barh(y=sorted_products.index, width=sorted_products.product_id)
plt.xlabel('Number of Sales')  # Label for the x-axis
plt.ylabel('Products')          # Label for the y-axis
plt.title('Top 10 Most Selling Products')  # Title of the plot

# # Adding values for each bar
# for index, value in zip(sorted_products.index, sorted_products.product_id):
#     plt.text(value, index, str(value))  # Adding the value next to each bar

plt.show()
```

### Pertanyaan 2:

```python
import matplotlib.pyplot as plt

# Sort the products by sales in descending order
sorted_products = most_profitable_product.sort_values(by='price', ascending=False).head(10)

# Create a horizontal bar plot
plt.barh(y=sorted_products.index, width=sorted_products.price)
plt.xlabel('Number of Sales')  # Label for the x-axis
plt.ylabel('Products')          # Label for the y-axis
plt.title('Top 10 Most Profitable Products')  # Title of the plot

# Adding values for each bar
# for index, value in zip(sorted_products.index, sorted_products.price):
#     plt.text(value, index, str(value))  # Adding the value next to each bar

plt.show()
```

**Insight:**
- Berdasarkan most selling product, bed_bath_table merupakan category name yang paling laris
- Berdasarkan most profitable product, health_beauty merupakan category name yang paling menguntungkan

## Conclusion
