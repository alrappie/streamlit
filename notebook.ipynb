## Data Wrangling

### Gathering Data

```python
customer_data = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\customers_dataset.csv')
order_items_data = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\order_items_dataset.csv')
products_data = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\products_dataset.csv')
product_translation = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\product_category_name_translation.csv')
order_data = pd.read_csv('A:\Pemrograman\Python\Belajar\Dicoding\streamlit\data\orders_dataset.csv')
```

```python
merged_data1 = pd.merge(order_items_data, products_data, how = 'inner',left_on= 'product_id',right_on= 'product_id')
merged_data2 = pd.merge(order_data, customer_data, how = 'inner',left_on= 'customer_id',right_on= 'customer_id')
merged_data3 = pd.merge(merged_data1,merged_data2, how='inner',left_on='order_id',right_on='order_id')
merged_data = pd.merge(merged_data3,product_translation,how='inner',left_on='product_category_name',right_on='product_category_name')
merged_data
```

**Insight:**
- Untuk menjawab 2 pertanyaan diatas diperlukan data-data yang dapat memberikan informasi seperti: harga kategori produk, jumlah pembelian suatu produk, dan kota dari masing-masing customer
- Setelah menentukan data yang digunakan maka dilakukan merged agar data dapat diolah ke proses lebih lanjut

### Cleaning Data

```python
data_cleaned = merged_data.dropna()
data_cleaned
```

```python
data_cleaned = data_cleaned.drop(columns=['product_name_lenght','seller_id','product_name_lenght','product_description_lenght','product_photos_qty','product_weight_g','product_length_cm','product_height_cm','product_width_cm','customer_zip_code_prefix','product_category_name','freight_value','shipping_limit_date','customer_state','order_approved_at',	'order_delivered_carrier_date',	'order_delivered_customer_date',	'order_estimated_delivery_date'])
data_cleaned
```

```python
data_cleaned.info()
```

```python
data_cleaned.order_purchase_timestamp = pd.to_datetime(data_cleaned.order_purchase_timestamp)
```

**Insight:**
- dikarenakan hanya 3.2% data saja yang missing, maka dapat dilakukan dropna
- dilakukan dropping column untuk mempermudah melakukan analysis
- mengubah tipe data pada order_purchase_timestamp        menjadi timestamp

## Exploratory Data Analysis (EDA)

### Explore ...

```python
data_cleaned.describe()
```

```python
max(data_cleaned.order_purchase_timestamp) # Max date
```

```python
most_selling_product = data_cleaned[data_cleaned.order_purchase_timestamp >= max(data_cleaned.order_purchase_timestamp) - pd.DateOffset(months=3)].groupby('product_category_name_english').agg({
    'product_id':'count',
    'price':'sum'
}).sort_values('product_id',ascending=False)
most_selling_product
```

```python
most_profitable_product = data_cleaned[data_cleaned.order_purchase_timestamp >= max(data_cleaned.order_purchase_timestamp) - pd.DateOffset(months=3)].groupby('product_category_name_english').agg({
    'product_id':'count',
    'price':'sum'
}).sort_values('price',ascending=False)
most_profitable_product
```

```python
top_city = data_cleaned[data_cleaned.order_purchase_timestamp >= '2018-01-01'].groupby('customer_city').agg({
    'product_id':'count',
}).sort_values('product_id',ascending=False)
top_city
```

**Insight:**
- berdasarkan data tersebut category produk yang paling dibeli dan menguntungkan dalam 3 bulan kebelakang adalah health beauty
- berdasarkan data tersebut kota yang memberikan kontribusi selama 1 tahun terakhir ini adalah sao paulo sebesar 10162

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
import matplotlib.pyplot as plt

# Sort the products by sales in descending order
sorted_products = most_selling_product.sort_values(by='product_id', ascending=False).head(5)

# Create a horizontal bar plot
plt.barh(y=sorted_products.index, width=sorted_products.product_id)
plt.xlabel('Number of Sales')  # Label for the x-axis
plt.ylabel('Products')          # Label for the y-axis
plt.title('Top 10 Most Selling Products')  # Title of the plot

# # Adding values for each bar
# for index, value in zip(sorted_products.index, sorted_products.product_id):
#     plt.text(value, index, str(value))  # Adding the value next to each bar

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sort the products by sales in descending order
sorted_products = most_profitable_product.sort_values(by='price', ascending=False).head(5)

# Create a horizontal bar plot
plt.barh(y=sorted_products.index, width=sorted_products.price)
plt.xlabel('Number of Sales')  # Label for the x-axis
plt.ylabel('Products')          # Label for the y-axis
plt.title('Top 10 Most Profitable Products')  # Title of the plot

# Adding values for each bar
# for index, value in zip(sorted_products.index, sorted_products.price):
#     plt.text(value, index, str(value))  # Adding the value next to each bar

plt.show()
```
